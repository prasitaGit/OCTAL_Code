```python
import networkx as nx
import matplotlib.pyplot as plt


# Defining a Class
class GraphVisualization:

    def __init__(self):

        # visual is a list which stores all
        # the set of edges that constitutes a
        # graph
        self.visual = []

    # addEdge function inputs the vertices of an
    # edge and appends it to the visual list
    def addEdge(self, a, b):
        temp = [a, b]
        
        self.visual.append(temp)
        
    def addListEdges(self, elist):
        self.visual = elist

    # In visualize function G is an object of
    # class Graph given by networkx G.add_edges_from(visual)
    # creates a graph with a given list
    # nx.draw_networkx(G) - plots the graph
    # plt.show() - displays the graph
    def visualize(self):
        G = nx.Graph()
        G.add_edges_from(self.visual)
        
        pos = nx.spring_layout(G, k=1.0, iterations=20)
        nx.draw_networkx(G,pos)
        plt.show()

# Driver code
color_map = []
G = GraphVisualization()
G.addEdge(0, 2)
G.addEdge(1, 2)
G.addEdge(1, 3)
G.addEdge(5, 3)
G.addEdge(3, 4)
G.addEdge(1, 0)
G.visualize()
```

```python
edgeList = []
for i in range(len(data_list[1][0].edge_index[0])):
    edgeList.append([data_list[1][0].edge_index[0][i].item(), data_list[1][0].edge_index[1][i].item()])
```

```python
edgeList
```

```python
import networkx as nx
color_map = []
G = nx.Graph()
for elem in edgeList:
    color='black'
    if ((elem[0] > 41 or elem[1] > 41) and not(elem[0] > 41 and elem[1] > 41)):
        color='red'
    G.add_edge(elem[0],elem[1],color=color)
    color_map.append("green")
    
```

```python
colors = nx.get_edge_attributes(G,'color').values()
colors
```

```python
pos = nx.spring_layout(G, k=1.0, iterations=20)
nx.draw_networkx(G,pos,edge_color=colors,node_color=color_map)
plt.show()
```
